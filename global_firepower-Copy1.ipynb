```python
import pandas as pd
import numpy as np
import seaborn as sns
import os

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv('Documents/Alex_practice/firepower.csv')
```

```python
df
```

```python
df.drop_duplicates()
pd.set_option('display.max_rows', None)
df
```

```python
# checking goverments active personnel
pd.set_option('display.max_columns', None)
df = df.sort_values(by=['Active Personnel'], inplace=False, ascending=False)
df
```

```python
df
```

```python
df.dtypes



```

```python
df['Defense Budget'] = df['Defense Budget'].fillna(0)
df['Defense Budget'] = df['Defense Budget'].astype('int64')

df['rate_off_active_pvm'] = df['rate_off_active_pvm'].astype('float')
df
```

```python
df['rate_off_active_pvm'] = df['Active Personnel']/df['Available Manpower']
df = df.sort_values(by=['rate_off_active_pvm'], inplace=False, ascending=False)
df
```

```python
df.dtypes
```

```python
# Scatter plot with budget vs gross
df = df.sort_values(by=['Defense Budget'], inplace=False, ascending=False)
df
plt.scatter(x=df['Active Personnel'],y=df['Defense Budget'])
plt.title('Active Personnel vs budget')
plt.xlabel('Active Personnel')
plt.ylabel('Defense Budget')
plt.show()
```

```python
df = df.sort_values(by=['rate_off_active_pvm'], inplace=False, ascending=False)
df
df.head()
```

```python
# Data between Defense Budget vs Active Personnel
sns.regplot(x='Defense Budget',y='rate_off_active_pvm',data=df, scatter_kws = {"color":"red"}, line_kws={"color":"blue"})
```

```python
print(df['Defense Budget'].corr(df['rate_off_active_pvm']))
```

```python
df1 = df[['Defense Budget','rate_off_active_pvm']]
df1


```

```python
correlation_matrix = df1.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True )
plt.title('Correlation matrix for Numeric Features')
plt.xlabel('Country Features')
plt.ylabel('Country Features')
plt.show()
```
